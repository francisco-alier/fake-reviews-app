```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
from datasets import load_dataset

dataset = load_dataset("theArijitDas/Fake-Reviews-Dataset")
```

```python
df = dataset["train"]
```

```python

plt.hist(df['rating'], bins=5, edgecolor='black')
plt.xlabel('Score')
plt.ylabel('Count')
plt.title('Histogram of Review Scores')
plt.show()
```
